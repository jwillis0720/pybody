```python
from sadie.airr import Airr
from sadie.antibody.genetable import VGeneTable
```

```python
v_gene = VGeneTable()
```

```python
airr_api = Airr("human")
catnap_heavy_results = airr_api.run_file(
    "../tests/integration/airr/fixtures/catnap_nt_heavy.fasta.gz"
```

```python
catnap_short_name = (
    catnap_heavy_results["sequence_id"]
    .str.split("_")
    .str.get(0)
    .to_frame()
    .rename({"sequence_id": "short_name"}, axis=1)
    .join(catnap_heavy_results.table)
)


for g, df in catnap_short_name.groupby("short_name"):
    with open("catnap_fasta/" + g + "_catnap_heavy.fasta", "w") as f:
        for _, x in df.iterrows():
            entry = v_gene.get_gene(x["v_call_top"], "human")
            v_gene_aa = entry[
                ["fwr1_aa", "cdr1_aa", "fwr2_aa", "cdr2_aa", "fwr3_aa", "cdr3_aa"]
            ].sum()
            short_name = x["short_name"]
            sequence_alignment = x["sequence_alignment_aa"]
            germline_alignment = x["germline_alignment_aa"]
            v_seq_alignment = x["v_sequence_alignment_aa"]
            v_germ_alignment = x["v_germline_alignment_aa"]
            v_igl_alignment = x["vdj_igl"]
            f.write(f">{short_name}_sequence\n{sequence_alignment}\n")
            f.write(f">{short_name}_germline\n{germline_alignment}\n")
            f.write(f">{entry.name}\n{v_gene_aa}\n")
            f.write(f">{short_name}_v_seq\n{v_seq_alignment}\n")
            f.write(f">{short_name}_v_germ\n{v_germ_alignment}\n")
            f.write(f">{short_name}_igl\n{v_igl_alignment}\n")
```
